```python
import nbody
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
b1 = nbody.PointMass('b1', 1, np.array([1., 0.]), np.array([0., 1.]))
b2 = nbody.PointMass('b2', 2, np.array([-1., 0.]), np.array([0., -1.]))

system = nbody.NBodySystem(b1, b2)
system.stationary()
sim = system.simulate('2s', '1ms', grav_const=10)
sim.head()
```

```python
fig, ax = plt.subplots()
ax.plot(sim['b1']['x1'], sim['b1']['x2'])
ax.plot(sim['b2']['x1'], sim['b2']['x2'])
ax.axis('equal')
plt.show()
```

```python
b3 = nbody.PointMass('b3', 10., np.array([0, 0]), np.array([0, 0]))
system3 = nbody.NBodySystem(b1, b2, b3)
system3.stationary()
sim3 = system3.simulate('10s', '1ms', grav_const=0.28)

fig2, ax2 = plt.subplots()
ax2.plot(sim3['b1']['x1'], sim3['b1']['x2'])
ax2.plot(sim3['b2']['x1'], sim3['b2']['x2'])
ax2.plot(sim3['b3']['x1'], sim3['b3']['x2'], color='red')
ax2.axis('equal')
plt.plot()
```
